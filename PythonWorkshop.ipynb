# Control Flows

Conditional Statements


> Conditional statements are like if-then statements which allow the programmer to execute code based on some Boolean condition. Python uses if, elif, and else.



```python
z = -15
```

```python
#@title
if z == 0:
  print(z, "is a zero.")
elif z > 0:
  print(z, "is positive")
elif z < 0:
  print(z, "is negative")
else:
  print(z, "is not real")
```

For Loop


> Loops in Python are a way to repeatedly execute a code.

> In a for loop, we specify the variable we want to use, the sequence we want to loop over, and use the *in* operator to link them together.





```python
for i in range (10):
  print(i)
```

Let us join a conditional statement with a for loop. 

```python
m = 5
```

```python
#@title
for i in range(10):
  if i < m:
    print(i, "is less than 5.")
  elif i == m:
    print(i, "is equal to 5.")
  else:
    print(i, "is greater than 5.")
```

# Machine Learning Algorithm

```python
X = df_water.drop('Potability', axis = 1)
X.head()
```

```python
y = df_water['Potability']
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
```

```python
from sklearn import metrics
from sklearn import tree

dtClf = tree.DecisionTreeClassifier(random_state = 123)
dtClf.fit(X_train, y_train)
y_pred2 = dtClf.predict(X_test)

y_pred2
```

```python
from sklearn.metrics import classification_report

y_pred2_dt= dtClf.predict(X_test)
print(classification_report(y_test, y_pred2_dt))
```

```python
metrics.confusion_matrix(y_test, y_pred2) #helps us define the performance of an algorithm.
```

True Positive: The predicted value matches the actual value.

True Negative: The predicted value matches the actual value.

False Positive (type 1 Error): The predicted value was falsely predicted
